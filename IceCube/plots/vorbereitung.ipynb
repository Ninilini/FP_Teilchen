```python
signal = pd.read_csv('signal.csv', sep=';')
signal.head()
```

Entfernung von Attributen, welche nicht in beiden Dateien vorkommen


```python
#Holen der Column-names
names_bg = list(bg.columns.values)
names_sg = list(signal.columns.values)

#Suche names, die nicht in Beiden sind

names_bg_cut_off = []
for i in names_bg:
    if i in names_sg:
        continue
    else:
        names_bg_cut_off.append(i)

names_sg_cut_off = []
for j in names_sg:
    if j in names_bg:
        continue
    else:
        names_sg_cut_off.append(j)
        
```

Entfernung der Simulierten Attribute:

```python
delet = np.array(['Corsica', 'MC', 'Weight', 'ID', 'Header', 'SPline' ])
for i in names_bg:
    for j in delet:
        if j in i:
            if i in names_bg_cut_off:
                continue
            else:
                names_bg_cut_off.append(i)
            if i in names_sg_cut_off:
                continue
            else:
                names_sg_cut_off.append(i)
```

```python
#Lösche columns
for i in names_bg_cut_off:
    del bg[i]

for j in names_sg_cut_off:
    del signal[j]
```
