```python
import numpy as np
import pandas as pd
```

```python
bg = pd.read_csv('background.csv', sep=';')
bg.head()
```

```python
signal = pd.read_csv('signal.csv', sep=';')
signal.head()
```

Entfernung von Attributen, welche nicht in beiden Dateien vorkommen


```python
#Holen der Column-names
names_bg = list(bg.columns.values)
names_sg = list(signal.columns.values)

#Suche names, die nicht in Beiden sind

names_bg_cut_off = []
for i in names_bg:
    if i in names_sg:
        continue
    else:
        names_bg_cut_off.append(i)

names_sg_cut_off = []
for j in names_sg:
    if j in names_bg:
        continue
    else:
        names_sg_cut_off.append(j)
        
#LÃ¶sche columns, die nicht in Beiden sind
for i in names_bg_cut_off:
    del bg[i]

for j in names_sg_cut_off:
    del signal[j]
```

```python
#Label schon als name vorhanden, deshalb nutze ich type als labelname
bg.insert(loc=0, column = 'type', value = 0)
signal.insert(loc=0, column = 'type', value = 1)
bg.head()
```
